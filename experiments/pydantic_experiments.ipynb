PYDANTIC

```python
from pydantic import BaseModel

class Player(BaseModel):
    name: str
    score: int

```

```python
player1 = Player(name="Alex", score=50)
print("Player 1:", player1)
```

```python
player1 = Player(name="Sam", score='fifty')
print("Player 1:", player1)
```

ONE AGENT WITH PERSONNA

```python
pirate_agent = Agent(
    "groq:qwen/qwen3-32b",
    name = "monkey-d-luffy",
    description="a great caption",
    system_prompt=(
        "You are Captain Monkey D Luffy. Who is on a mission to find one peice."
        "You must answer every question like a king of pirates."
        "Never drop your characcter."
    )
)
```

```python
pirate_agent
```
